```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df1 = df.copy()
df1["species"] = df1["species"].replace({"setosa": 0, "versicolor": 1, "virginica": 2})
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=101)
```

```python
X= df1.drop("species", axis=1)
```

```python
preds = kmeans.fit(X)
```

```python
preds = kmeans.labels_
x = X.values
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,7))
sns.scatterplot(x[preds == 0,0], x[preds == 0,1], label = "Setosa", color="red", s=100)
sns.scatterplot(x[preds == 1,0], x[preds == 1,1], label = "Versiclor", color="green", s=100)
sns.scatterplot(x[preds == 2,0], x[preds == 2,1], label = "Virginica", color="blue", s=100)
sns.scatterplot(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], label ="centroids", s=150, color="black")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Speal Length")
plt.ylabel("Sepal Width")
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x[preds == 0,0], x[preds == 0,2], label = "Setosa", color="red", s=100)
sns.scatterplot(x[preds == 1,0], x[preds == 1,2], label = "Versicolor", color="green", s=100)
sns.scatterplot(x[preds == 2,0], x[preds == 2,2], label = "Virginica", color="blue", s=100)
sns.scatterplot(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,2], label ="centroids", s=150, color="black")
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Speal Length")
plt.ylabel("Petal Length")
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x[preds == 0,2], x[preds == 0,3], label = "Setosa", color="red", s=100)
sns.scatterplot(x[preds == 1,2], x[preds == 1,3], label = "Versicolor", color="green", s=100)
sns.scatterplot(x[preds == 2,2], x[preds == 2,3], label = "Virginica", color="blue", s=100)
sns.scatterplot(kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,3], label ="centroids", s=150, color="black")
plt.title("Petal Legth vs Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
```
