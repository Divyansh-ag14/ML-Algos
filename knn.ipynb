```python
import pandas as pd
```

```python
df = pd.read_csv("knn_data.csv")
```

```python
df.head()
```

```python
def knn(k):
    name = input("Enter name : ")
    Age  = int(input("Enter age : "))
    Gender  = input("Enter gender (M/F) : ")
    
    en = {"M":0, "F": 1}
    df["Gender"] = df["Gender"].map(en)
    
    print("-----------------------------------------------------------------------------")
    
    if (Gender =="M"):
        Gender = 0
        s="he"
    else:
        Gender = 1
        s="she"
    
    age= list(df["Age"] - Age)
    age = [number ** 2 for number in age]
    
    gen = list(df["Gender"] - Gender)
    gen = [number ** 2 for number in gen]
    
    res = [age[i] + gen[i] for i in range(len(age))]
    res = [number ** 0.5 for number in res]
    
    df["Distance"] = res
    
    ans = df.nsmallest(k, ['Distance']).Sport.mode() #check the most frequent value 
    ans=list(ans)
    
    en = {0:"M", 1: "F"}
    df["Gender"] = df["Gender"].map(en)
    
    print(k, "nearest neighbours are :")
    df2=df.nsmallest(k, ['Distance'])
    display(df2)
    
    print("-----------------------------------------------------------------------------")
    print("Predicted sport is :",ans)
    print("-----------------------------------------------------------------------------")
    
    if(len(ans)==3):
        print("Well,",name,"can play either Cricket or Football with : ",end="")
        print(df2[((df2["Sport"]=="Cricket") | (df2["Sport"]=="Football"))]["Name"].values.tolist())
        print("-----------------------------------------------------------------------------")
        print("Or,",s,"can chill with : ",end="")
        print(df2[df2["Sport"]=="Neither"]["Name"].values.tolist())
        print("-----------------------------------------------------------------------------")
    
    elif(ans[0]=="Neither"):
        print("Seems like,",name,"is boring. So,",s,"can get along with : ",end="")
        print(df2[df2["Sport"] == ans[0]]["Name"].values.tolist())
        print("-----------------------------------------------------------------------------")
    
    else:
        print("So,",name,"can play with : ",end="")
        print(df2[df2["Sport"] == ans[0]]["Name"].values.tolist())
        print("-----------------------------------------------------------------------------")
    
    df.drop(["Distance"], axis=1, inplace=True)
    
    choice = input("Do you wish to make another prediction (Y/N) ? ")
    if(choice=="Y"):
        k= int(input("Enter value of K : "))
        knn(k)
    else:
        print("Okay :(:")
    
   
        
    
```
